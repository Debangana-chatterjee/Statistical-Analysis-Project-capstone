```python
import pandas as pd
import statsmodels.api as sm
```

```python
#Loading the data from Excel file
data= pd.read_excel("Dataset_1.xlsx")
data
```

```python
# Filter for 2-bedroom houses as specified in the user's end 
filtered_data = data[(data['NO_OF_BEDROOMS'] == 2) & data['FLOOR_SIZE'].notna() &
                     data['NO_OF_BATHROOMS'].notna() & data['PRICE_SQFT'].notna()]

filtered_data.describe(), filtered_data.head()
```

```python
# Defining the independent variables and dependent variable
X = filtered_data[['FLOOR_SIZE', 'NO_OF_BATHROOMS', 'PRICE_SQFT']]
y = filtered_data['PRICE']
```

```python
# Adding a constant to the model for the intercept

X = sm.add_constant(X)
```

```python
# Performing the multiple regression analysis
model = sm.OLS(y, X).fit()
```

```python
# Summary of the regression analysis

model_summary = model.summary()
model_summary
```

R-squared: 0.985, suggesting that approximately 98.5% of the variance in price can be explained by floor size, number of bathrooms, and price per square foot.

FLOOR_SIZE: ùëù < 0.001 highly significant.

NO_OF_BATHROOMS: ùëù = 0.044 also significant, though less  than the other variables.

PRICE_SQFT: ùëù < 0.001 highly significant.

Since all independent variables show significant p-values (less than 0.05), we reject the null hypothesis. This suggests that floor size, number of bathrooms, and price per square foot each have a statistically significant effect on price at a 95% confidence level.
